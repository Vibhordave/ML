## Without Scikit Learn

```python
import pandas as pd
import matplotlib.pyplot as plt

penguins = pd.read_csv("penguins.csv")
penguins_db=pd.DataFrame()
penguins_db['flipper_length_mm']=penguins['flipper_length_mm']
penguins_db['body_mass_g']=penguins['body_mass_g']
# penguins=penguins['flipper_length_mm']+penguins['body_mass_g']

penguins=penguins_db
penguins.head()
```

```python
import seaborn as sns

feature_name = "flipper_length_mm"
target_name = "body_mass_g"
data, target = penguins[[feature_name]], penguins[target_name]

ax = sns.scatterplot(
    data=penguins, x=feature_name, y=target_name, color="black", alpha=0.5
)
ax.set_title("Body Mass as a function of the Flipper Length")
```

```python
def linear_model_flipper_mass(flipper_length, weight_flipper_length, intercept_body_mass):
    """Linear model of the form y = a * x + b"""
    body_mass = weight_flipper_length * flipper_length + intercept_body_mass
    return body_mass
```

```python
import numpy as np

weight_flipper_length = 45
intercept_body_mass = -5000 # Initially taken

flipper_length_range = np.linspace(data.min(), data.max(), num=300)
predicted_body_mass = linear_model_flipper_mass(
    flipper_length_range, weight_flipper_length, intercept_body_mass
)
flipper_length_range
```

```python
label = "{0:.2f} (g / mm) * flipper length + {1:.2f} (g)"

ax = sns.scatterplot(
    data=penguins, x=feature_name, y=target_name, color="black", alpha=0.5
)
ax.plot(flipper_length_range, predicted_body_mass)
_ = ax.set_title(label.format(weight_flipper_length, intercept_body_mass))
```

## With Scikit Learn

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
data=imputer.fit_transform(data)
penguins['flipper_length_mm']=data
print(data)

penguins['body_mass_g']=imputer.fit_transform(penguins[['body_mass_g']])
target=penguins['body_mass_g']
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(data, target)
```

```python
# Parameters are stored in coef_ and intercept_
weight_flipper_length = linear_regression.coef_[0] # w value
intercept_body_mass = linear_regression.intercept_ # b value

flipper_length_range = np.linspace(data.min(), data.max(), num=300)
predicted_body_mass = (
        weight_flipper_length * flipper_length_range + intercept_body_mass
)
sns.scatterplot(penguins,x=feature_name, y=target_name, color="black", alpha=0.5)
plt.plot(flipper_length_range, predicted_body_mass)
_ = plt.title("Model using LinearRegression from scikit-learn")
```

```python
from sklearn.metrics import mean_squared_error

inferred_body_mass = linear_regression.predict(data)
model_error = mean_squared_error(target, inferred_body_mass)
print(f"The mean squared error of the optimal model is {model_error:.2f}")
```
