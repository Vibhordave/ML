```python
# Kernal density estimation helps us estimate the probability density from the given data set
#It is done in non parametric way
# In this method we add a kernal which is basically a gaussian or a trangular or a cosine distribution centered at each point and then add them up
# we start with seaborn's univariate plot
cars = sns.load_dataset('mpg').dropna()

sns.kdeplot(cars.horsepower, fill=True) # fill shades the the area beneath the graph, bw is bandwidth and bw is basically proportional to variance however it is not working perfectly
plt.show()
# If we want CDF instead of PDF
sns.kdeplot(cars.horsepower, cumulative=True)
plt.show()

# For a 2D contour KDE plot
sns.kdeplot(data=cars,x='horsepower',y='mpg',n_levels=20,fill=True,shade_lowest=False,cbar=True); # Basically x and y are the axes given , n_levels increases the number of rings
plt.show()

```

```python
# distplot - distribution plot
# combines histogram kdeplot and rugplot that is all the distribution plot all into 1
# rugplot is 1D representation of the data
sns.distplot(cars.weight,rug=True) # kde=False removes KDE plot and hist=False removes histogram
plt.show()

# increasing the number of bins=20 increases the number of bars
sns.distplot(cars.weight,bins=20) # bins can also take input as list of values and each value is taken as a start point for the next bin and end point of the last
plt.show()

from scipy.stats import skewnorm
sns.distplot(cars.weight,fit=skewnorm,kde=False) # fits the given plot to the data
plt.show()

# Further we will see that we can add arguments specific to kde histogram or rug plots
sns.distplot(cars.weight, kde=False, hist_kws={'alpha': 1, 'histtype': 'stepfilled'});
plt.show()

sns.distplot(cars.weight, kde_kws={'lw': 5});
plt.show()

sns.distplot(cars.weight, rug=True, rug_kws={'height': 0.15, 'color': 'xkcd:navy'});
plt.show()

plt.clf()
plt.figure(figsize=(8,5))
sns.distplot(cars.weight,
             bins=20,
             kde_kws={"lw": 8, 'color':'xkcd:bluish green'},
             hist_kws={'alpha': 0.3, 'color':'gray'}
            )

sns.despine(left=True)
plt.xlabel('Weight (lbs)', fontsize=16, labelpad=15)
plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
# boxplot is the plot where median 25 percentile and 75 percentile points are marked
# distance between 25 to 75 percentile is called Inner Quartile Range(IQR)
# now we make whiskers which goes to both left and right till the last data points which is  <= 1.5*IQR and the left out points are marked with diamond
# these diamonds are actually not quite useful and hence clearly it helps in telling the usefulness of the datapoints
# it is mainly used when we need to split into groups and it is also called categorical distribution plot

# cars = cars[cars.cylinders.isin([4,6,8])] # for removing some excess useless data , uncomment before running
sns.boxplot(cars.mpg); # makes a general box plot but not quite useful
plt.show()

sns.boxplot(x='origin', y='mpg', data=cars); # actually categorizes mpg on the basis of origin
plt.show()

sns.boxplot(x='origin', y='mpg', hue='cylinders', data=cars, hue_order=[4,8,6]); # adds an extra category to check via hue
plt.show()

# we can use fliersize to adjust size diamond and whis value which is by default 1.5 also linewidth in the graph and other coloring factor and also hue order i.e. in which order is the hue columns categories are shown(As seen above)
```

```python
# Violin plot
# basically a box plot merged with two mirror image KDE plot forms a violin plot
# Categorical distribution plot
# instead of  outliers the KDE plot is used and we will observe low KDE value around the outliers

cars = cars[cars.cylinders.isin([4,6,8])]

sns.violinplot(cars.displacement)
plt.show()

# categorical distribution
sns.violinplot(y=cars.displacement,x=cars.cylinders)
plt.show()

# 3rd category
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars)
plt.show()

# overlapping 2 categories
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True) # can be used only if hue has only 2 categories
plt.show()

# if we wanted to show the quartiles instead of the box plot inside the violin
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,inner='quartiles') # can be used only if hue has only 2 categories
plt.show()

# we can actually scale our violin plot so that we can actually observe the number of observations present
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,scale='count',scale_hue=False) # can be used only if hue has only 2 categories
plt.show() # we actually see here that with the number of cylinder as 6 the no. of cars produced is very less

# order can be used to have a particular order
# linewidth can be used for increasing linewidth
# bw is used for bandwidth
```

```python
# Get population change in US by querying for US data
df_us = px.data.gapminder().query("country == 'United States'")
px.bar(df_us, x='year', y='pop')

# Create a stacked bar with more customization
df_tips = px.data.tips()
px.bar(df_tips, x='day', y='tip', color='sex', title='Tips by Sex on Each Day',
       labels={'tip': 'Tip Amount', 'day': 'Day of the Week'})

# Place bars next to each other
px.bar(df_tips, x="sex", y="total_bill",
       color='smoker', barmode='group')

# Display pop data for countries in Europe in 2007 greater than 2000000
df_europe = px.data.gapminder().query("continent == 'Europe' and year == 2007 and pop > 2.e6")
fig = px.bar(df_europe, y='pop', x='country', text='pop', color='country')
# Put bar total value above bars with 2 values of precision
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
# Set fontsize and uniformtext_mode='hide' says to hide the text if it won't fit
fig.update_layout(uniformtext_minsize=8)
# Rotate labels 45 degrees
fig.update_layout(xaxis_tickangle=-45)
```

```python
# Plot histogram based on rolling 2 dice
dice_1 = np.random.randint(1,7,5000)
dice_2 = np.random.randint(1,7,5000)
dice_sum = dice_1 + dice_2
# bins represent the number of bars to make
# Can define x label, color, title
# marginal creates another plot (violin, box, rug)
fig = px.histogram(dice_sum, nbins=11, labels={'value':'Dice Roll'},
                   title='5000 Dice Roll Histogram', marginal='violin',
                   color_discrete_sequence=['green'])

fig.update_layout(
    xaxis_title_text='Dice Roll',
    yaxis_title_text='Dice Sum',
    bargap=0.2, showlegend=False
)

# Stack histograms based on different column data
df_tips = px.data.tips()
px.histogram(df_tips, x="total_bill", color="sex")
```

```python
# A box plot allows you to compare different variables
# The box shows the quartiles of the data. The bar in the middle is the median 
# The whiskers extend to all the other data aside from the points that are considered
# to be outliers
df_tips = px.data.tips()
# We can see which sex tips the most, points displays all the data points
px.box(df_tips, x='sex', y='tip', points='all')

# Display tip sex data by day
px.box(df_tips, x='day', y='tip', color='sex')

# Adding standard deviation and mean
fig = go.Figure()
fig.add_trace(go.Box(x=df_tips.sex, y=df_tips.tip, marker_color='blue',
                     boxmean='sd'))

# Complex Styling
df_stocks = px.data.stocks()
fig = go.Figure()
# Show all points, spread them so they don't overlap and change whisker width
fig.add_trace(go.Box(y=df_stocks.GOOG, boxpoints='all', name='Google',
                     fillcolor='blue', jitter=0.5, whiskerwidth=0.2))
fig.add_trace(go.Box(y=df_stocks.AAPL, boxpoints='all', name='Apple',
                     fillcolor='red', jitter=0.5, whiskerwidth=0.2))
# Change background / grid colors
fig.update_layout(title='Google vs. Apple',
                  yaxis=dict(gridcolor='rgb(255, 255, 255)',
                             gridwidth=3),
                  paper_bgcolor='rgb(243, 243, 243)',
                  plot_bgcolor='rgb(243, 243, 243)')
```

```python
# Violin Plot is a combination of the boxplot and KDE
# While a box plot corresponds to data points, the violin plot uses the KDE estimation
# of the data points
df_tips = px.data.tips()
px.violin(df_tips, y="total_bill", box=True, points='all')

# Multiple plots
px.violin(df_tips, y="tip", x="smoker", color="sex", box=True, points="all",
          hover_data=df_tips.columns)

# Morph left and right sides based on if the customer smokes
fig = go.Figure()
fig.add_trace(go.Violin(x=df_tips['day'][ df_tips['smoker'] == 'Yes' ],
                        y=df_tips['total_bill'][ df_tips['smoker'] == 'Yes' ],
                        legendgroup='Yes', scalegroup='Yes', name='Yes',
                        side='negative',
                        line_color='blue'))
fig.add_trace(go.Violin(x=df_tips['day'][ df_tips['smoker'] == 'No' ],
                        y=df_tips['total_bill'][ df_tips['smoker'] == 'No' ],
                        legendgroup='Yes', scalegroup='Yes', name='No',
                        side='positive',
                        line_color='red'))
```
